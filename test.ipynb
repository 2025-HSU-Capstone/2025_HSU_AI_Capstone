```python
import pandas as pd

df = pd.read_csv("/Users/heohyeonjun/Desktop/2025 AI_Capstone_Project/AI_System_Capstone/data/recipes1.csv")
df.head()
```

```python
df_subset = df[["RCP_TTL", "CKG_MTRL_CN", "CKG_IPDC"]].copy()
df_subset.columns = ["title", "ingredients", "intro"]

df_subset.head()
```

```python
import re

def clean_ingredient_text(text):
    if pd.isna(text):
        return ""
    
    # ✅ 0. 제어 문자 제거 (ASCII 0~31 + 127)
    text = re.sub(r"[\x00-\x1F\x7F]", "", text)
    
    # ✅ 1. [재료] 제거, 구분자 정리
    text = re.sub(r"\[.*?\]", "", text)
    text = text.replace("|", ",")

    # ✅ 2. 재료 나누기
    items = [item.strip() for item in text.split(",")]

    cleaned = []
    for item in items:
        # ✅ 3. 숫자+단위 제거 (1/2개, 200g 등)
        item = re.sub(r"\d+(\/\d+)?\s*[가-힣]*", "", item)
        item = re.sub(r"[½¼¾⅓⅔⅛]", "", item)
        # ✅ 4. 영어 제거
        item = re.sub(r"[a-zA-Z]", "", item)
        # ✅ 5. 남은 특수문자 제거
        item = re.sub(r"[^\uAC00-\uD7A3]", "", item)

        item = item.strip()
        if item:
            cleaned.append(item)

    return ", ".join(sorted(set(cleaned)))

df_subset["ingredients"] = df_subset["ingredients"].apply(clean_ingredient_text)
df_subset["intro"] = df_subset["intro"].fillna("")

df_subset.head()
```

```python
# ✅ 5. 문서 생성 (RAG용 문서)
def make_document(row):
    return f"{row['title']}\n\n[ingredients]\n{row['ingredients']}\n\n[intro]\n{row['intro']}"
```

```python
df_subset["document"] = df_subset.apply(make_document, axis=1)
```

```python
print(df_subset["document"].iloc[0])
```

```python
# 컬럼명을 영어로 맞춰줬다면 (선택적으로 다시 한 번 확인)
df_subset.columns = ["title", "ingredients", "intro", "document"]

# CSV 파일로 저장
df_subset.to_csv("recipes1_new.csv", index=False, encoding="utf-8-sig")

print("✅ 저장 완료 → recipes1_new.csv")
```
